_**Празький град**_


`Ця фортеця в Празі служить резиденцією президента Чехії. Це найбільша резиденція лідера країни на планеті. Крім того, Празький град претендує на звання найбільшого замку у Світі. Побудований ще в IX-X столітті, замок височіє над Прагою, відразу привертаючи увагу гостей міста.`

_**Порохова вежа**_


`Навіть на тлі численних старовинних будівель центру Праги Порохова вежа виділяється своєю незвичайною архітектурою. Це досить строгий будинок може здатися занадто похмурим, адже є зразком готичного стилю. Усередині Порохової вежі розташовується виставка фотографій, проте туристів більше приваблює можливість піднятися 186 сходинками на оглядовий майданчик. Звідси, з висоти 44 метрів відкривається красивий вид на центр міста.`

_**Староміська площа**_


`Цю старовинну празьку площу можна назвати справжньою виставкою найпопулярніших архітектурних стилів. На Староміської площі можна побачити оригінальні, прекрасно збережені будівлі в стилі пишного бароко, вишуканого ренесансу, романтичного рококо, суворої готики. У середині XX століття Староміська площа стала пішохідною зоною. Сьогодні тут можна покататися в старовинній кареті, побачити знамениті Празькі куранти, просто прогулятися, милуючись старовинними будівлями.`
